```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Realizar a leitura do meu arquivo csv
df = pd.read_csv('C:/Users/dsadm/Desktop/Desenvolvimento_Back-End_2_semestre_ADS/Aula 3 e 4/base_de_dados/monitoramento_tempo.csv')
```

```python
# exibindo as primeiras linhas do dataframe
df.head() # exibe as 4 primeiras linhas do dataframe
```

```python
# importa a biblioteca datetime
import datetime # Biblioteca que permite a transformação da data que ea um object
# em datetime
df['data'] = pd.to_datetime(df['data']) # converti a data de object para datetime
plt.figure(figsize=(15,8))
plt.plot(df['data'],df['temperatura'])
plt.xlabel('Data e hora')
plt.ylabel('Temperatura')
```

```python
fig = plt.figure(figsize=(15,8)) # crio uma variável fig para armazenar o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # define o tamanho dos eixos do gráfico
eixo.plot(df['data'],df['temperatura']/10)
eixo.set_title('Temperatura no momento',fontsize=25) # título do gráfico
eixo.legend(['temperatura'],loc='lower right',fontsize = 15)
eixo.set_xlabel('Ano',fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=20)
```
